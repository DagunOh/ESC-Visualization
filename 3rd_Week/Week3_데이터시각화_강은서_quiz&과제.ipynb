## Quiz1

```python
gapminder = pd.read_csv('gapminder.csv')
gapminder = gapminder[gapminder['Year']==1970]
gapminder = gapminder.reset_index(drop=True)
gapminder
```

```python
from bokeh.models import HoverTool
from bokeh.models import ColumnDataSource
from bokeh.models.widgets import Panel, Tabs
```

```python
source = ColumnDataSource(gapminder)


hover = HoverTool(tooltips=[('Country', '@Country'), ('Fertility', '@fertility'),
                            ('Life Expectancy', '@life'), ('GDP', '@gdp')])

p1 = figure(plot_width=800, plot_height=400, tools=[hover],
              x_axis_label='Fertility', y_axis_label='Life Expectancy', title='Life Expectancy in 1970')
mapper = CategoricalColorMapper(factors=['South Asia','Europe & Central Asia','Middle East & North Africa',
                                         'Sub-Saharan Africa', 'America' ,'East Asia & Pacific' ],
                                palette='Spectral6')
p1.circle('fertility', 'life',legend_field='region',source=source, size=5, color={'field':'region', 'transform':mapper})
p1.legend.location = 'bottom_left'


p2 = figure(plot_width=800, plot_height=400, tools='box_select, reset',
              x_axis_label='Petal Length', y_axis_label='Sepal Length', title='Petal Length vs. Sepal Length')
p2.circle('fertility', 'life',legend_field='region',source=source, size=5, color={'field':'region', 'transform':mapper})
p2.legend.location = 'bottom_left'
```

```python
tab1 = Panel(child=p1, title='Hover')
tab2 = Panel(child=p2, title='Box select')

tabs = Tabs(tabs=[tab1, tab2])
show(tabs)
```

- energy: represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy.

- key
- liveness: 라이브 정도, 청중의 유무를 판단해 0~1사이 값으로 표현
- mode : 조성
- speechiness: 음악속의 단어 등을 감지해내는 수치. 0~1사이 값으로 표현
- tempo
- time_signature : 박자 가늠하는정도...?
- valence: 음악 속의 부정적인 감정과 긍정적인 감정의 정도를 표현하는 수치. 0~1사이 값으로 표현

#### 4. 장르 및 키별 pairplot

```python
gnr = list(spotify['genre'].unique())
kkk= list(spotify['key'].unique())
```

```python
@interact(genre = gnr, key= kkk)

def update(genre, key):
    spotify2 = spotify[(spotify['genre']==genre) & (spotify['key']==key)]
    a = sns.pairplot(spotify2[['danceability','energy','loudness','tempo','popularity']],
                     diag_kind='hist', plot_kws = {'alpha':0.5, 's':80, 'edgecolor':'k'},
                     diag_kws = {'alpha':0.3}, height=6)
    plt.show()
```

눈에 띄는 상관관계가 있는 변수는 energy와 loudness정도.

```python
from bokeh.models.widgets import Panel, Tabs
from bokeh.layouts import gridplot
from bokeh.layouts import column
from bokeh.layouts import row
from bokeh.models import CategoricalColorMapper
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool
```

```python
kpop_girl = spotify[(spotify['artist_name']=='TWICE') | (spotify['artist_name']=='BLACKPINK')|
                   (spotify['artist_name']=='Red Velvet')|(spotify['artist_name']=='Apink')|
                   (spotify['artist_name']=='ITZY')|(spotify['artist_name']=='IU')]
```

```python
kpop_boy = spotify[(spotify['artist_name']=='BTS') | (spotify['artist_name']=='EXO')|
                   (spotify['artist_name']=='Zico')|(spotify['artist_name']=='TAEYANG')|
                   (spotify['artist_name']=='SHINee')|(spotify['artist_name']=='WINNER')]
```

```python
p1= figure(plot_width=800, plot_height=400, x_axis_label='Tempo', y_axis_label='Popularity',
             title='Tempo vs. Popularity: K-pop Girl Group', tools='box_select, reset')
source = ColumnDataSource(kpop_girl)
mapper = CategoricalColorMapper(factors=['TWICE', 'BLACKPINK', 'Red Velvet','Apink','ITZY','IU'],
                                palette=['lightcoral','indianred', 'tomato','pink','hotpink','purple']
                               )
p1.circle('valence', 'popularity',size=10, color={'field':'artist_name', 'transform':mapper},
            legend_field='artist_name', source=source)
p1.legend.location = 'bottom_left'

source = ColumnDataSource(kpop_girl)
hover = HoverTool(tooltips=[('Artist', '@artist_name'), ('Track name', '@track_name'),
                            ('Tempo', '@tempo'), ('Popularity','@popularity')])
p2 = figure(plot_width=800, plot_height=400, tools=[hover],
              x_axis_label='Tempo', y_axis_label='Popularity')
p2.circle('tempo', 'popularity', source=source, color={'field':'artist_name', 'transform':mapper},
            legend_field='artist_name',size=10)
p2.legend.location = 'bottom_left'


p3 = figure(plot_width=800, plot_height=400, x_axis_label='Tempo', y_axis_label='Popularity',
             title='Tempo vs. Popularity: K-pop Boy Group', tools='box_select, reset')
source = ColumnDataSource(kpop_boy)
mapper = CategoricalColorMapper(factors=['BTS','EXO','Zico','TAEYANG','SHINee','WINNER'],
                                palette=['turquoise','teal','deepskyblue','dodgerblue','slateblue','darkviolet']
                               )
p3.circle('valence', 'popularity',size=10, color={'field':'artist_name', 'transform':mapper},
            legend_field='artist_name', source=source)
p3.legend.location = 'bottom_left'

source = ColumnDataSource(kpop_boy)
hover = HoverTool(tooltips=[('Artist', '@artist_name'), ('Track name', '@track_name'),
                            ('Tempo', '@tempo'), ('Popularity','@popularity')])
p4 = figure(plot_width=800, plot_height=400, tools=[hover],
              x_axis_label='Tempo', y_axis_label='Popularity')
p4.circle('tempo', 'popularity', source=source, color={'field':'artist_name', 'transform':mapper},
            legend_field='artist_name', size=10)
p4.legend.location = 'bottom_left'
```

```python
tab1 = Panel(child=column(p1,p2), title='Girl Group')
tab2 = Panel(child=column(p3,p4), title='Boy Group')

tabs = Tabs(tabs=[tab1, tab2])
show(tabs)
```
