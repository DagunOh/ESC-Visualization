```python
import matplotlib as mpl 
import matplotlib.pyplot as plt ;plt.rcParams['font.family'] = 'NanumGothic'
import pandas as pd

covid2 = pd.read_csv("C:/Users/njj06/Desktop/quiz1.csv")
```

```python
fig, ax = plt.subplots()
ax.grid(True, color='gray',zorder=0, axis='y')
ax.bar(covid2['일'], covid2['해외유입'], label='해외유입', zorder=2)
ax.bar(covid2['일'], covid2['국내발생'], bottom=covid2['해외유입'], label='국내발생',zorder=2)
ax.set_xticklabels(covid2['일'], rotation=90)
ax.set_title('감염경로에구분에 따른 신규확진자 현황')
ax.legend()
plt.show()
```
