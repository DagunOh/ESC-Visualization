```python
#라이브러리
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects #text edge color 추가용
import pandas as pd
import warnings

%matplotlib inline
warnings.filterwarnings(action='ignore') #경고 무시하기
```

```python
#글꼴
plt.rcParams['font.family'] = 'NanumGothic'
```

# Quiz1

```python
data1 = pd.read_csv('data/quiz1.csv')
data1
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.grid(True, color='gray', linestyle='-', linewidth=0.5, axis='y', zorder=0)
ax.bar(data1['일'], data1['해외유입'], label='해외유입', color='hotpink', zorder=2)
ax.bar(data1['일'], data1['국내발생'], bottom=data1['해외유입'], label='국내발생', color='dodgerblue', zorder=2)
ax.set_ylim([0, 70])
ax.set_title('감염경로구분에 따른 신규확진자 현황', fontsize=18)

ax.set_ylabel('(명)', rotation=0)
ax.yaxis.set_label_coords(-0.05,1.02)

ax.legend(bbox_to_anchor=(0.55, -0.08), fontsize=15)
for idx in range(7):
    ax.text(data1['일'][idx], data1['해외유입'][idx]+1, data1['해외유입'][idx], color='hotpink', fontsize=13, fontweight='bold',
            path_effects=[path_effects.Stroke(linewidth=3, foreground='white'), path_effects.Normal()])
    ax.text(data1['일'][idx], data1['해외유입'][idx]+data1['국내발생'][idx]+1, data1['국내발생'][idx],
            color='dodgerblue', fontsize=13, fontweight='bold')
plt.show()
```

# Quiz2

```python
data2 = pd.read_csv('data/quiz2.csv')
data2
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.grid(True, color='gray', linestyle='-', linewidth=0.5, axis='y', zorder=0)
ax.bar(data2['일'], data2['누적 확진환자'], label='누적 확진환자', zorder=2)
ax.set_ylabel('누적 환진환자(명)', rotation=0, color='C0', fontweight='bold')
ax.yaxis.set_label_coords(-0.05,1.02)

ax2 = ax.twinx()
ax2.plot(data2['일'], data2['일 확진환자'], label='일 확진환자', lw=4, color='orange', marker='o',
         ms=10, mec='white', mew=2, mfc='orange', zorder=2)
ax2.set_ylim([0, 100])
ax2.set_ylabel('일 환진환자(명)', rotation=0, color='orange', fontweight='bold')
ax2.yaxis.set_label_coords(1,1.05)
for idx in range(7):
    ax2.text(data2['일'][idx], data2['일 확진환자'][idx]+3, data2['일 확진환자'][idx], color='red',
             fontsize=13, fontweight='bold',
             path_effects=[path_effects.Stroke(linewidth=3, foreground='white'), path_effects.Normal()])
    
fig.suptitle('일일 및 누적 확진환자 추세', fontsize=20)
fig.legend(bbox_to_anchor=(0.55, 0.08), fontsize=12)
plt.show()
```
