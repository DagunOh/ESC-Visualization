```python
import matplotlib as mpl 
import matplotlib.pyplot as plt ;plt.rcParams['font.family'] = 'NanumGothic'
import pandas as pd

covid2 = pd.read_csv("C:/Users/njj06/Desktop/quiz1.csv")
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.grid(True, color='gray',zorder=0, axis='y')
ax.bar(covid2['일'], covid2['해외유입'], label='해외유입', zorder=2, width=0.5)
ax.bar(covid2['일'], covid2['국내발생'], bottom=covid2['해외유입'], label='국내발생',zorder=2, width=0.5)
ax.set_xticklabels(covid2['일'])
ax.set_title('감염경로에구분에 따른 신규확진자 현황')
ax.legend(bbox_to_anchor=(0.55, -0.08))
plt.show()
```

```python
covid3 = pd.read_csv("C:/Users/njj06/Desktop/quiz2.csv")
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.grid(True, zorder=0, axis='y')
ax.bar(covid3['일'], covid3['누적 확진환자'], label='누적확진환자', width=0.5, zorder=3)
ax2=ax.twinx()
ax2.plot(covid3['일'], covid3['일 확진환자'], label='일확진환자', color='orange', marker='o')
ax2.set_ylim((0,90))
ax.set_ylabel('누적확진환자')
ax2.set_ylabel('일확진횐자')
ax.set_title('일일 및 누적 확진환자 추세')
fig.legend(bbox_to_anchor=(0.55, 0.09))
plt.show()
```
