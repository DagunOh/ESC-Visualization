```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'AppleGothic'
```

```python
quiz1 = pd.read_csv("./ESC-Visualization/1st_Week/quiz1.csv")
quiz2 = pd.read_csv("./ESC-Visualization/1st_Week/quiz2.csv")

quiz1

```

```python
fig, ax =plt.subplots(figsize=(30,10))

ax.grid(True, color='gray', linewidth=1,axis='y',zorder=0)
ax.bar(quiz1['일'], quiz1['해외유입'], label = '해외유입',color='deeppink', width=0.5, zorder=5)
ax.bar(quiz1['일'],quiz1['국내발생'], label = '국내발생', bottom = quiz1['해외유입'], color='royalblue', width=0.5,zorder=10)
ax.set_title('감염경로구분에 따른 신규확진자 현황', fontsize=20, loc='left')
ax.legend(bbox_to_anchor=(0.55,0.007),fontsize=15)
 
for x, y in enumerate( list(quiz1['해외유입']) ):
        txt = "%d" % y
        ax.text(x, y, txt, fontsize=20, color='white', 
                    horizontalalignment='center', verticalalignment='bottom',zorder=15 )
        
plt.show()
```

```python
fig, ax =plt.subplots(figsize=(20,10))
ax.set_ylim([0, 14000])
ax.grid(True, color='gray', linewidth=1,axis='y',zorder=0)
ax.bar(quiz2['일'],quiz2['누적 확진환자'], label = '누적 확진환자', color='royalblue', width=0.4)
ax.set_ylabel('누적 확진환자',rotation=0, color='blue', fontsize=15)
ax.yaxis.set_label_coords(-0.02,1.03)

ax.text(6, 13816, '13816', fontsize=20, color='blue', 
                    horizontalalignment='center', verticalalignment='bottom',zorder=15,fontweight='bold')
ax2 = ax.twinx()
ax2.set_ylim([0, 90])
ax2.plot(quiz2['일'], quiz2['일 확진환자'], label= '일 확진환자', color='orange', marker='o', linewidth=4, markersize=10)        
ax2.set_ylabel('일일 확진환자',rotation=0, color='red', fontsize=15)
ax2.yaxis.set_label_coords(1.02,1.05)

for i in range(7):
    ax2.text(quiz2['일'][i],quiz2['일 확진환자'][i]+2,quiz2['일 확진환자'][i], fontsize=15, color= 'orange', horizontalalignment='center', verticalalignment='bottom')

fig.suptitle('일일 및 누적 확진환자 추세', fontsize=20)
fig.legend(bbox_to_anchor=(0.55,0.007),fontsize=15)

plt.show()
```
