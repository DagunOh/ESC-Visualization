### 연도별 초중고등생 1인당 사교육비(만원)

```python
edu2
```

```python
fig, ax = plt.subplots()
ax.plot(edu2['연도'], edu2['초등학교'], label='초등학교', marker='.')
ax.plot(edu2['연도'], edu2['중학교'], label='중학교', marker='.')
ax.plot(edu2['연도'], edu2['고등학교'], label='고등학교', marker='.')
ax.grid(True, color='gray',zorder=0, axis='y')
ax.set_xlabel('연도')
ax.set_ylabel('사교육비(단위:만원)')
ax.set_xticks(edu2['연도'])
ax.set_title('연도별 1인당 월평균 사교육비', fontsize=15)
ax.legend()
plt.show
```

```python
fig, ax = plt.subplots(1, 3, sharex=True, sharey=True, figsize = (10,5))
ax[0].plot(edu2['연도'], edu2['초등학교'], label='초등학교', marker='.')
ax[1].plot(edu2['연도'], edu2['중학교'], label='중학교', marker='.', c='orange')
ax[2].plot(edu2['연도'], edu2['고등학교'], label='고등학교', marker='.', c='green')
ax[0].set_title('초등학교')
ax[1].set_title('중학교')
ax[2].set_title('고등학교')
ax[0].set_xticks(range(2009,2020,2))
plt.show
```

```python
fig, ax = plt.subplots()
ax.bar(edu1['시도별'], edu1['고등학교'])
ax.set_xticklabels(edu1['시도별'], rotation=90)
ax.set_xlabel('시도명')
ax.set_ylabel('사교육비(만원)')
ax.set_title('2019년 고등학생 1인당 월평균 사교육비', fontsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
x = np.arange(len(edu1['시도별']))
ax.bar(x-0.2, edu1['초등학교'], label='초등학교', width=.2)
ax.bar(x-0.0, edu1['중학교'], label='중학교', color='orange', width=.2)
ax.bar(x+0.2, edu1['고등학교'], label='고등학교', color='green', width=.2)
ax.set_xticks(x)
ax.set_xticklabels(edu1['시도별'])
ax.set_xlabel('시도명')
ax.set_ylabel('사교육비(만원)')
ax.legend()
ax.set_title('2019년 초중고등학생 1인당 월평균 사교육비', fontsize=17)
plt.show()
```

```python
pop = pd.read_csv("C:/Users/njj06/Desktop/인구.csv", encoding = 'cp949')
```

### 2019년 인구 동태 자료

```python
pop.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.pie(pop.iloc[:,3], labels=pop['행정구역별'], autopct='%1.1f%%')
ax.set_title('2019년 행정구역별 혼인건수 비율', fontsize = 15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(pop['행정구역별'], pop['출생건수'], label='출생건수', width=0.5, zorder=3)
ax2=ax.twinx()
ax2.plot(pop['행정구역별'], pop['합계출산율'], label='출산율', color='orange', marker='o')
ax.set_ylabel('출생건수(명)')
ax2.set_ylabel('출산율(가임여성 1명당)')
ax.set_xticklabels(pop['행정구역별'], rotation=90)
ax.set_title('2019년 인구 동태', fontsize=15)
fig.legend()
plt.show()
```
