```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus']=False
# 한글 깨지지 않도록 글씨체 설정
plt.rcParams['font.family'] = 'SeoulHangang'
```

# 1. 2020년 6월 7호선 역별 일평균 승하차 인원 Bar-Chart로 나타내기

```python
data=pd.read_excel('6월 역별 시간대별 이용인원(홈페이지).xlsx')
```

```python
data
```

```python
line7=data[data['호선']=='7호선']
line7
```

```python
line7day=line7[['날짜','역번호','역명','구분','합 계']]
line7day
```

```python
# 역별 일평균 승차/하차 인원
line7day_mean=line7day.groupby(['구분','역번호','역명'],as_index=False)[['합 계']].mean()
line7day_mean
```

## 2020년 6월 7호선 역별 일평균 승차 인원

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.grid(True, color='gray', linewidth=1, axis='y', alpha=0.2, zorder=0)
ax.bar(line7day_mean[line7day_mean['구분']=='승차']['역명'],
       line7day_mean[line7day_mean['구분']=='승차']['합 계'],
      color='darkolivegreen')
ax.set_ylim([0,80000])
ax.set_xticklabels(line7day_mean[line7day_mean['구분']=='승차']['역명'],rotation=90)
ax.set_ylabel('(명)',rotation='horizontal')
ax.yaxis.set_label_coords(-0.02,1.04)
ax.set_title('2020년 6월 7호선 역별 일평균 승차량', fontsize=25)
plt.show()
```

## 2020년 6월 7호선 역별 일평균 하차 인원

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.grid(True, color='gray', linewidth=1, axis='y', alpha=0.2, zorder=0)
ax.bar(line7day_mean[line7day_mean['구분']=='하차']['역명'],
       line7day_mean[line7day_mean['구분']=='하차']['합 계'],
      color='darkolivegreen')
ax.set_ylim([0,80000])
ax.set_xticklabels(line7day_mean[line7day_mean['구분']=='하차']['역명'],rotation=90)
ax.set_ylabel('(명)',rotation='horizontal')
ax.yaxis.set_label_coords(-0.02,1.04)
ax.set_title('2020년 6월 7호선 역별 일평균 하차량', fontsize=25)
plt.show()
```

## 2020년 6월 7호선 역별 일평균 승하차 인원

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.grid(True, color='gray', linewidth=1, axis='y', alpha=0.2, zorder=0)
ax.bar(line7day_mean[line7day_mean['구분']=='승차']['역명'],
       line7day_mean[line7day_mean['구분']=='승차']['합 계'],
      color='darkolivegreen')
ax.bar(line7day_mean[line7day_mean['구분']=='하차']['역명'],
       line7day_mean[line7day_mean['구분']=='하차']['합 계'],
       bottom=line7day_mean[line7day_mean['구분']=='승차']['합 계'],
      color='darkolivegreen')
ax.set_ylim([0,80000])
ax.set_xticklabels(line7day_mean[line7day_mean['구분']=='하차']['역명'],rotation=90,fontsize=9)
ax.set_ylabel('(명)',rotation='horizontal')
ax.yaxis.set_label_coords(-0.02,1.04)
ax.set_title('2020년 6월 7호선 역별 일평균 승하차량', fontsize=25)
plt.show()
```

# 2. 2020년 6월 7호선 주요 역 시간대별 평일 승하차인원 Time-Series Graph

```python
line7t=line7.drop(['호선','합 계'],axis=1)
line7t
```

```python
# 주말,공휴일 삭제
line7twd=line7t[~((line7t['날짜']=='2020-06-06')
                   &(line7t['날짜']=='2020-06-07')
                   &(line7t['날짜']=='2020-06-13')
                   &(line7t['날짜']=='2020-06-14')
                   &(line7t['날짜']=='2020-06-20')
                   &(line7t['날짜']=='2020-06-21')
                   &(line7t['날짜']=='2020-06-27')
                   &(line7t['날짜']=='2020-06-28'))]
line7twd
```

```python
line7twd_mean=line7twd.groupby(['구분','역번호','역명'],as_index=False)[line7twd.columns[3:]].mean()
line7twd_mean
```

## 가산디지털단지역 (업무밀집지역)

```python
gasan=line7twd_mean[line7twd_mean['역명']=='가산디지털단지']
gasan
```

```python
gasant=gasan.transpose().drop(['역번호','역명'])
gasant=gasant.rename(columns=gasant.iloc[0])
gasant=gasant.drop(gasant.index[0])
gasant
```

## 가산디지털단지역 시간대별 평일 승하차인원

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.grid(True, color='gray', linewidth=1, axis='y', alpha=0.2, zorder=0)
ax.plot(gasant.index,gasant['승차'],label='승차')
ax.plot(gasant.index,gasant['하차'],label='하차')
ax.set_xticklabels(gasant.index,rotation=90)
ax.set_xlabel('시간대')
ax.set_ylabel('(명)', rotation='horizontal')
ax.yaxis.set_label_coords(-0.02,1.04)
ax.set_title('7호선 가산디지털단지역 시간대별 평일 승하차인원',fontsize=20)
ax.legend()
plt.show()
```

오전 peak time에는 하차인원이, 오후 peak time에는 승차인원이 주로 분포

## 신중동역 (상업시설, 주거 골고루 분포하는 지역)

```python
sjd=line7twd_mean[line7twd_mean['역명']=='신중동']
sjdt=sjd.transpose().drop(['역번호','역명'])
sjdt=sjdt.rename(columns=sjdt.iloc[0])
sjdt=sjdt.drop(sjdt.index[0])
sjdt
```

## 신중동역 시간대별 평일 승하차인원

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.grid(True, color='gray', linewidth=1, axis='y', alpha=0.2, zorder=0)
ax.plot(sjdt.index,sjdt['승차'],label='승차')
ax.plot(sjdt.index,sjdt['하차'],label='하차')
ax.set_xticklabels(sjdt.index,rotation=90)
ax.set_xlabel('시간대')
ax.set_ylabel('(명)', rotation='horizontal')
ax.yaxis.set_label_coords(-0.02,1.04)
ax.set_title('7호선 신중동역 시간대별 평일 승하차인원',fontsize=20)
ax.legend()
plt.show()
```

Peak time의 승하차인원이 오전, 오후 모두 고르게 분포
