# 우리나라 산업 구조

## Data1) 1970년대~2010년대 산업별 명목 부가가치 비중 변화

```python
data1 = pd.read_csv('indust.csv')
```

```python
data1
```

### 1-1. Bar Chart

```python
fig, ax = plt.subplots()
ax.set_ylim([0, 100])
ax.bar(data1['시기'], data1['1차산업'], label='1차 산업', width=5, color='cornflowerblue')
ax.bar(data1['시기'], data1['2차산업'], bottom=data1['1차산업'],
       label='2차 산업', width=5, color='lightcoral')
ax.bar(data1['시기'], data1['3차산업'], bottom=data1['2차산업']+data1['1차산업'],
       label='3차 산업', width=5, color='mediumaquamarine')
ax.set_xticks(data1['시기'])
ax.set_title('산업구조 변화')
ax.legend(loc='upper left')
plt.show()
```

### 1-2. Time Series

```python
fig, ax = plt.subplots()
ax.plot(data1['시기'], data1['1차산업'], label='1차 산업',
        color='cornflowerblue', marker='.')
ax.plot(data1['시기'], data1['2차산업'], label='2차 산업',
        color='lightcoral', marker='.')
ax.plot(data1['시기'], data1['3차산업'], label='3차 산업',
        color='mediumaquamarine', marker='.')
ax.set_xticks(data1['시기'])
ax.set_title('산업구조 변화')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, sharex=True, figsize=(11,4))

ax[0].set_ylim([0, 100])
ax[0].bar(data1['시기'], data1['1차산업'], label='1차 산업', width=5, color='cornflowerblue')
ax[0].bar(data1['시기'], data1['2차산업'], bottom=data1['1차산업'],
          label='2차 산업', width=5, color='lightcoral')
ax[0].bar(data1['시기'], data1['3차산업'], bottom=data1['2차산업']+data1['1차산업'],
          label='3차 산업', width=5, color='mediumaquamarine')
ax[0].set_title('Bar Chart')

ax[1].plot(data1['시기'], data1['1차산업'], label='1차 산업',
           color='cornflowerblue', marker='.')
ax[1].plot(data1['시기'], data1['2차산업'], label='2차 산업',
           color='lightcoral', marker='.')
ax[1].plot(data1['시기'], data1['3차산업'], label='3차 산업',
           color='mediumaquamarine', marker='.')
ax[1].set_title('Time Series')

fig.legend(['1차 산업', '2차 산업', '3차 산업'], loc='center left')
plt.show()
```

### 1-3. Pie Chart(2010년대 산업 구조)

```python
fig, ax = plt.subplots()
ax.pie(data1.iloc[4, 1:], labels=data1.columns[1:], counterclock=False,
       colors=['cornflowerblue', 'lightcoral', 'mediumaquamarine'], explode=(0, 0.1, 0))
ax.set_title('2010년대 산업 구조')
plt.show()
```

## Data2) 제조업(2차 산업) 권역별 출하액, 종사자, 사업체 비중

```python
data2 = pd.read_csv('manuf.csv')
```

```python
data2
```

### 2-1. Bar Chart

```python
fig, ax = plt.subplots(figsize=(4,5))
ax.set_ylim([0, 100])
ax.bar(data2.columns[1:], data2.iloc[0, 1:], label='수도권', width=0.5, color='navy')
ax.bar(data2.columns[1:], data2.iloc[1, 1:],
       bottom=data2.iloc[0,1:],
       label='영남권', width=0.5, color='steelblue')
ax.bar(data2.columns[1:], data2.iloc[2, 1:],
       bottom=data2.iloc[0,1:]+data2.iloc[1,1:],
       label='충청권', width=0.5, color='deepskyblue')
ax.bar(data2.columns[1:], data2.iloc[3, 1:],
       bottom=data2.iloc[0,1:]+data2.iloc[1,1:]+data2.iloc[2,1:],
       label='호남권', width=0.5, color='cornflowerblue')
ax.bar(data2.columns[1:], data2.iloc[4, 1:],
       bottom=data2.iloc[0,1:]+data2.iloc[1,1:]+data2.iloc[2,1:]+data2.iloc[3,1:],
       label='기타', width=0.5, color='paleturquoise')
ax.set_title('권역별 출하액, 종사자, 사업체 비중')
fig.legend(loc='lower center', ncol=5)
plt.show()
```

### 2-2. Pie Chart

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0, 0].pie(data2.iloc[:, 1], labels=data2.iloc[:, 0])
ax[0, 1].pie(data2.iloc[:, 2], labels=data2.iloc[:, 0])
ax[0, 2].pie(data2.iloc[:, 3], labels=data2.iloc[:, 0])
plt.show()
```

이거 왜 계속 에러 뜨는지 모르겠음ㅠㅠ 결국 그냥 따로 따로 세개 그렸다

```python
fig, ax = plt.subplots()
ax.pie(data2.iloc[:, 1], labels=data2.iloc[:, 0], colors=['navy', 'steelblue', 'deepskyblue', 'cornflowerblue', 'paleturquoise'])
ax.set_title('권역별 출하액 비중')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(data2.iloc[:, 2], labels=data2.iloc[:, 0], colors=['navy', 'steelblue', 'deepskyblue', 'cornflowerblue', 'paleturquoise'])
ax.set_title('권역별 종사자 수 비중')
plt.show()
```
