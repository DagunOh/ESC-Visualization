```python
import matplotlib.font_manager as fm
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt ; plt.rcParams['font.family']='malgun'
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
```

```python
movie = pd.read_csv('movies.csv')
movie = movie.dropna(axis=0)
epl = pd.read_csv('epl.csv')
```

```python
fig, ax = plt.subplots(2,2, figsize=(13, 13), sharex=True)
sns.countplot(x='club', data=epl, hue='big_club', ax=ax[0,0])
sns.boxplot(x='club', y='market_value', data=epl, hue='big_club', ax=ax[0,1])
sns.barplot(x='club', y='market_value', data=epl, ax=ax[1,0])
sns.pointplot(x='club', y='market_value', data=epl, hue='big_club', capsize=0.5, join=False, ax=ax[1,1])
ax[0,0].set_title('Count Plot')
ax[0,1].set_title('Box Plot')
ax[1,0].set_title('Bar Plot')
ax[1,1].set_title('Point Plot')
ax[1, 0].set_xticklabels(labels=epl['club'].unique(), rotation=90)
ax[1, 1].set_xticklabels(labels=epl['club'].unique(), rotation=90)
fig.suptitle('Several Plots')
plt.show()
```

```python
sns.set_style('whitegrid')
a = sns.pairplot(movie[['dir_prev_bfnum', 'box_off_num', 'screening_rat']], hue='screening_rat',
                diag_kind='hist', plot_kws = {'alpha':0.5, 's':80, 'edgecolor':'k'},
                diag_kws = {'alpha':0.3}, palette='RdBu', markers=['o', 's', 'D', '+'], height=5, aspect=1.2)
a.fig.suptitle('Audience', x=0.45, y=1.05, fontsize=24)
plt.show()
```

## +) Bike Sharing Dataset

```python
london = pd.read_csv('london_merged.csv')
london["timestamp"] = pd.to_datetime(london["timestamp"])

london = london.set_index("timestamp")

london["year"] = london.index.year
london["hour"] = london.index.hour
london["day_of_month"] = london.index.day
london["day_of_week"]  = london.index.dayofweek
london["month"] = london.index.month

london.head()
```

```python
july = london[london['month']==7]
july.head()
```

```python
a = sns.catplot(x='month', y='cnt', data=london, kind='bar', palette='coolwarm', height=5, aspect=2)
a.ax.set_title('Month')
plt.show()
```

```python
a = sns.catplot(x='hour', y='cnt', data=july, kind='bar', palette='Blues', height=5, aspect=2)
a.ax.set_title('July_24hours')
plt.show()
```

#### 2016년 data 추출하여 빨간 날과 평일 구분

```python
y2016 = london[london['year']==2016]
y2016[['is_holiday']] = y2016[['is_holiday']].astype('bool')
y2016[['is_weekend']] = y2016[['is_weekend']].astype('bool')
red = y2016[(y2016['is_holiday']==True) | (y2016['is_weekend']==True)]
weekdays = y2016[(y2016['is_holiday']==False) & (y2016['is_weekend']==False)]
```

```python
y2016['red'] = (y2016['is_holiday']==True) | (y2016['is_weekend']==True)
```

```python
a = sns.catplot(x='month', y='cnt', data=y2016, kind='bar', hue='red', palette='coolwarm', height=4.8, aspect=2, ci=None)
a.ax.set_ylabel('count')
a.ax.set_xlabel('month')
a.ax.set_ylim(0,1800)
a.ax.set_title('Weekdays or Not', size=18)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9.5,5))
sns.barplot(x='month', y='cnt', data=weekdays, color='lightskyblue', ci=None)
sns.barplot(x='month', y='cnt', data=red, color='lightsalmon', ci=None)
ax.set_title('Weekdays or Not', size=18)
ax.set_ylabel('count')
ax.set_ylim(0, 1800)
plt.show()
```

```python
fig,ax = plt.subplots(4,1, figsize=(18,25))

sns.pointplot(x='hour', y='cnt', data=london, ax=ax[0])
sns.pointplot(x='hour', y='cnt', data=london, hue='is_holiday', ax=ax[1], palette='coolwarm')
sns.pointplot(x='hour', y='cnt', data=london, hue='is_weekend', ax=ax[2], palette='coolwarm')
sns.pointplot(x='hour', y='cnt', data=london, hue='season', ax=ax[3], palette='coolwarm')
```
