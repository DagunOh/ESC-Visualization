```python
import pandas as pd
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
import seaborn as sns
import warnings
plt.rcParams['font.family'] = 'NanumGothic'
warnings.filterwarnings(action='ignore')
```

# Quiz1

```python
movie = pd.read_csv('../data/movies.csv')
movie = movie.dropna(axis=0)
epl = pd.read_csv('../data/epl.csv')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))

a= sns.countplot(x='club', data=epl, hue='big_club', ax=ax[0, 0])
a.set(xticks=[])
a.set_title('Count Plot')

b= sns.boxplot(x='club', y='market_value', data=epl, hue='big_club', ax=ax[0, 1])
b.set(xticks=[])
b.set_title('Boxplot')

c= sns.barplot(x='club', y='market_value', data=epl, ax=ax[1,0])
c.set_xticklabels(labels=epl['club'].unique(), rotation=90)
c.set_title('Bar Plot')

d=sns.pointplot(x='club', y='market_value', data=epl, hue='big_club', capsize=0.5, join=False, ax=ax[1,1])
d.set_xticklabels(labels=epl['club'].unique(), rotation=90)
d.set_title('Point Plot')

fig.suptitle('Several Plots', fontsize=15)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.color_palette("cubehelix", 8)
plt.rcParams['font.family'] = 'NanumGothic'
a = sns.pairplot(data=movie[['dir_prev_bfnum', 'box_off_num', 'screening_rat']], hue='screening_rat',
                 diag_kind='hist', plot_kws = {'alpha':0.5, 's': 80, 'edgecolor':'k'}, diag_kws = {'alpha':0.3},
                 markers=['o', 's', 'D', '+'], size = 6)
a.fig.suptitle('관객 수', x=0.4, y=1)
plt.show()
```
